УГАДАЙКА

```python
import random
print('***Добро пожаловать в числовую угадайку***')
l = 1
print('Укажите правую границу')
r = int(input())
num = random.randint(l, r)
count = 0

def is_valid(s):
  return s.isdigit() and 1 <= int(s) <= r

while True:
  print('Угадайте число')
  n = input()
  is_valid(n)
  count += 1
  if is_valid(n) == False:
    print(f'А может быть все-таки введем целое число от 1 до {r}?')
    continue
  else:
    n = int(n)
  if n < num:
    print('Ваше число меньше загаданного, попробуйте еще разок')
  elif n > num:
    print('Ваше число больше загаданного, попробуйте еще разок')
  elif n == num:
    print(f'Вы угадали с {count} попытки, поздравляем!')
    print(f'Попробовать снова? y/n')
    a = input()
    if a == 'y':
      continue
    elif a == 'n':
      break


```
